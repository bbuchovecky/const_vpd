# Get paths

````python
## Constant VPD experiment
path_fixed = '/glade/scratch/czarakas/archive/OFFL0000_PI_fixedVPD_v2/lnd/hist/'

## Control VPD experiment
path_control = '/glade/scratch/czarakas/archive/OFFL0000_PI_v13/lnd/hist/'
````

````python
fixed = xr.open_dataset(path_fixed+'OFFL0000_PI_fixedVPD_v2.clm2.h0.0049-02-01-00000.nc')
control = xr.open_dataset(path_control+'OFFL0000_PI_v13.clm2.h0.0049-02-01-00000.nc')
````

# Dictionaries and constants

````python
case = {'fix':fixed, 'ctl':control}

lnd_variable = [
    'gssunln', # sunlit leaf stomatal conductance at local noon [umol/m2/s]
    'fctr',    # canopy transpiration [W/m2]
    'btran2',  # root zone soil wetness factor [1]
    'btranmn', # daily minimum of transpiration beta factor [1]
    'tlai',    # total projected leaf area index [m2/m2]
    'gpp',     # gross primary production [gC/m2/s]
    'tsa',     # 2m air temperature [K]
]

unit = {
    'gssunln':'[$\mu$mol H$_2$O m$^{-2}$ s$^{-1}$]',
    'fctr':'[W m$^{-2}$]',
    'btran2':'[1]',
    'btranmn':'[1]',
    'tlai':'[m$^2$ m$^{-2}$]',
    'gpp':'[gC m$^{-2}$ s$^{-1}$]',
    'tsa':'[K]',
    'tranlai':'[(W m$^{-2}$) (m$^2$ m$^{-2}$)$^{-1}$]'
}

subplot_kws = {'projection': ccrs.Robinson()}
cbar_kwargs = {'orientation':'vertical', 'fraction':0.025}
````

````python
vardict = {}

for cc in case.keys():
    vardict[cc] = {}
    for vv in lnd_variable:
         vardict[cc][vv] = case[cc][vv.upper()]
````

# Single month

````python
it = 0
var = 'gssunln'
(vardict['fix'][var].isel(time=it)-vardict['ctl'][var].isel(time=it)).plot(robust=True)
````

````python
it = 0
var = 'fctr'
(vardict['fix'][var].isel(time=it)-vardict['ctl'][var].isel(time=it)).plot(robust=True)
````

````python
it = 0
var = 'tlai'
(vardict['fix'][var].isel(time=it)-vardict['ctl'][var].isel(time=it)).plot(robust=True)
````

# Time average

````python
var = 'gssunln'
(vardict['fix'][var].mean(dim='time')-vardict['ctl'][var].mean(dim='time')).plot(cmap=cmo.balance,
                                                                                 vmin=-2e6, vmax=2e6, extend='both',
                                                                                 cbar_kwargs={**cbar_kwargs, **{'label':var.upper()+' '+unit[var]}}, 
                                                                                 subplot_kws=subplot_kws,
                                                                                 transform=ccrs.PlateCarree())
````

````python
var = 'fctr'
(vardict['fix'][var].mean(dim='time')-vardict['ctl'][var].mean(dim='time')).plot(cmap=cmo.balance,
                                                                                 vmin=-50, vmax=50, extend='both',
                                                                                 cbar_kwargs={**cbar_kwargs, **{'label':var.upper()+' '+unit[var]}},
                                                                                 subplot_kws=subplot_kws,
                                                                                 transform=ccrs.PlateCarree())
````
